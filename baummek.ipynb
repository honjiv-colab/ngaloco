```python
# @title cetmek
# scbaru.py
# This script runs a specific task for 5 minutes, pauses for a "random job" for 10 minutes, and then repeats the cycle.
# The script is self-contained and uses subprocess to run its own "worker" function.

import time
import subprocess
import sys
import os

# --- Configuration ---
# Duration for the worker task in seconds (5 minutes)
WORKER_DURATION = 5 * 60
# Duration for the "random job" in seconds (10 minutes)
RANDOM_JOB_DURATION = 10 * 60

def worker_task():
    """
    This function contains the code for the main task that needs to run for 5 minutes.
    You can replace the content of this function with your actual script's logic.
    """
    print("Worker: The worker task is starting now.")
    start_time = time.time()

    # This is the main loop for the task. It will run until the orchestrator
    # terminates the process or a local timeout is reached.
    try:
        while True:
            elapsed_time = time.time() - start_time
            # Optional: Add a local check to ensure it doesn't run longer than expected.
            if elapsed_time >= WORKER_DURATION:
                print(f"Worker: Worker task is self-terminating after {elapsed_time:.2f} seconds.")
                break

            # --- Your task code goes here ---
            # For demonstration, we'll just print a message every few seconds.
            print(f"Worker: Doing some work... (Elapsed: {elapsed_time:.2f}s / {WORKER_DURATION}s)")
            time.sleep(5)
            # ---------------------------------

    except KeyboardInterrupt:
        print("Worker: Worker process was manually interrupted.")

    print("Worker: The worker task has finished.")


def orchestrator():
    """
    This is the main function that orchestrates the execution loop.
    It runs the worker task as a subprocess, waits, and then runs the random job.
    """
    print("Orchestrator: Script started. Press Ctrl+C to stop.")

    while True:
        try:
            # --- Step 1: Run the worker task for the specified duration ---
            print(f"\nOrchestrator: Starting the worker task for {WORKER_DURATION / 60} minutes.")

            # Start the worker task as a subprocess. We pass '--worker' as an argument
            # to tell the subprocess to run the worker_task function.
            # Fix: Using sys.argv[0] to reliably get the script's path instead of __file__.
            process = subprocess.Popen([sys.executable, os.path.abspath(sys.argv[0]), '--worker'])

            # Wait for the specified duration (5 minutes)
            time.sleep(WORKER_DURATION)

            print(f"Orchestrator: {WORKER_DURATION / 60} minutes are up. Terminating worker task.")

            # Terminate the worker subprocess
            process.terminate()

            # Wait for the subprocess to gracefully exit. Add a timeout to prevent hanging.
            try:
                process.wait(timeout=5)
            except subprocess.TimeoutExpired:
                print("Orchestrator: Worker process did not terminate gracefully, killing it.")
                process.kill()

            print("Orchestrator: Worker task terminated successfully.")

            # --- Step 2: Perform the "random job" ---
            print(f"\nOrchestrator: Doing a 'random job' for {RANDOM_JOB_DURATION / 60} minutes.")

            # This is a placeholder for your "random job".
            # You can replace this sleep with a function call or another subprocess.
            # Example:
            # run_your_other_script()

            time.sleep(RANDOM_JOB_DURATION)

            print("Orchestrator: 'Random job' finished.")

        except KeyboardInterrupt:
            print("\nOrchestrator: Main script was manually interrupted. Exiting.")
            break
        except Exception as e:
            print(f"\nOrchestrator: An error occurred: {e}")
            print("Orchestrator: Restarting the loop after a short pause.")
            time.sleep(10)


if __name__ == "__main__":
    # Check if the script is running in worker mode
    if len(sys.argv) > 1 and sys.argv[1] == '--worker':
        worker_task()
    else:
        # If not, run the orchestrator
        orchestrator()

```
